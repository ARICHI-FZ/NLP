# Scrapping from BBC Arabic

```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.bbc.com/arabic'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extraire les titres des actualités
for headline in soup.find_all('h3'):
    print(headline.text.strip())
```

```python
# Tous les liens des articles sur la page d'accueil
articles = soup.find_all('a', href=True)
for article in articles:
    link = article['href']
    # BBC utilise des liens relatifs pour les articles, il faut les compléter
    if link.startswith('/arabic'):
        full_link = f'https://www.bbc.com{link}'
        print(article.text.strip(), full_link)
```

```python
# Choisissez un lien d'article (nous prenons le premier pour l'exemple)
article_url = 'https://www.bbc.com' + articles[0]['href']

# une nouvelle requête pour obtenir le contenu de l'article
response = requests.get(article_url)
article_soup = BeautifulSoup(response.text, 'html.parser')

# Extraction du texte de l'article
article_text = ''
for paragraph in article_soup.find_all('p'):
    article_text += paragraph.text
print(article_text)
```

```python
# Function to tokenize Arabic text
def tokenize_arabic(text):
    # Arabic script is in Unicode range 0600 to 06FF
    pattern = re.compile(r'[\u0600-\u06FF]+')
    # Find all words based on the Arabic script pattern
    words = pattern.findall(text)
    return words
```

```python
import nltk
from nltk.corpus import stopwords

# Function to remove Arabic stopwords from a list of tokens
def remove_arabic_stopwords(tokens):
    # Get the list of Arabic stopwords
    arabic_stopwords = set(stopwords.words('arabic'))
    
    # Remove stopwords from the list of tokens
    filtered_tokens = [token for token in tokens if token not in arabic_stopwords]
    
    return filtered_tokens 
```

```python
cleaned_text_f = remove_arabic_stopwords(tokens)
print(cleaned_text_f)
```

```python
def normalize_arabic_tokens(tokens):
    normalized_tokens = []
    for token in tokens:
        # Normalize each token
        token = re.sub(r"(مبرووو+ك)", "مبروك", token)
        token = re.sub(r"(هههه+)", "ههه", token)
        token = re.sub(r'(.)\1+', r'\1\1', token)  # Reduce character repetition
        # ... additional normalization rules here
        
        normalized_tokens.append(token)
    
    return normalized_tokens
```

```python
# Now, apply the normalization to each token in the list 'cleaned_text_f'
normalized_tokens = normalize_arabic_tokens(cleaned_text_f)

print(normalized_tokens)
```

```python
from collections import Counter

def discretize_tokens_by_frequency(tokens, frequency_threshold):
    # Count the frequency of each word in the list
    word_counts = Counter(tokens)

    # Discretize into high and low frequency bins
    high_frequency_words = [word for word, count in word_counts.items() if count > frequency_threshold]
    low_frequency_words = [word for word, count in word_counts.items() if count <= frequency_threshold]

    return high_frequency_words, low_frequency_words
```

```python
frequency_threshold = 2

high_freq_words, low_freq_words = discretize_tokens_by_frequency(normalized_tokens, frequency_threshold)

print("High frequency words:", high_freq_words)
print("Low frequency words:", low_freq_words)
```

# Stemming

 Stemming algorithmically cuts off prefixes and suffixes to reduce a word to its stem form. It's a heuristic process that removes common word endings for a particular language, which might not result in an actual word but a token that represents a group of words with similar meanings.

```python
def simple_arabic_stemmer(word):
    prefixes = ['ال', 'و', 'ف', 'ب', 'ك', 'ل', 'لل']
    suffixes = [ 'ي', 'ات', 'ان', 'ون', 'ين', 'تين', 'تان', 'وا']
                
    for prefix in prefixes:
        if word.startswith(prefix):
            word = word[len(prefix):]
            break
    
    for suffix in suffixes:
        if word.endswith(suffix):
            word = word[:-len(suffix)]
            break
    
    return word

def stem_arabic_tokens(tokens):
    stemmed_tokens = []
    for token in tokens:
        stemmed_token = simple_arabic_stemmer(token)
        stemmed_tokens.append(stemmed_token)
    return stemmed_tokens
```

```python
stemmed_tokens = stem_arabic_tokens(normalized_tokens)
print(stemmed_tokens)
```

# Lemmatization

Lemmatization reduces words to their base or dictionary form (lemma) by considering the word's part of speech and meaning in context. It's a more sophisticated approach that involves understanding the morphology of the word, often requiring a detailed dictionary of the language.

```python
import stanza

# Ensure the Arabic model is downloaded
stanza.download('ar')

# Function to lemmatize Arabic text (from pre-tokenized input) using Stanza
def lemmatize_arabic_text_from_tokens(tokens):
    # Reconstruct the text from pre-tokenized input
    text = ' '.join(tokens)
    
    # Initialize the Stanza pipeline (including POS tagging for accurate lemmatization)
    nlp = stanza.Pipeline(lang='ar', processors='tokenize,pos,lemma', use_gpu=False)

    # Process the reconstructed text
    doc = nlp(text)

    # Extract lemmatized tokens
    lemmatized_tokens = [word.lemma for sent in doc.sentences for word in sent.words]

    return lemmatized_tokens
```

```python
lemmatized_tokens = lemmatize_arabic_text_from_tokens(normalized_tokens)
print(lemmatized_tokens)
```
